```python
import numpy as np
import pandas as pd
import math
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import urllib.request
```

```python
df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.info()
```

```python
df["user_id"] = df["user_id"].astype("string")
df["user_name"] = df["user_name"].astype("string")
df["review_title"] = df["review_title"].astype("string")
df["review_content"] = df["review_content"].astype("string")
df["img_link"] = df["img_link"].astype("string")
df["product_link"] = df["product_link"].astype("string")
```

```python
#Removing white spaces from 'product_id' column
df['product_id'] = df['product_id'].str.strip()
```

```python
#Fixing the strings in 'category' column
print(df["category"][0],"\n",df["category"][10],"\n",df["category"][100])
```

```python
df_category = df['category'].str.split('|', expand=True).rename(columns={0:'category_1',
1:'category_2', 2:'category_3', 3:'category_4', 4:'category_5'})
df_category.head()
```

DATA CLEANING AND PREPARATION

```python
#Checking the missing value in category
df_category.isna().sum()
```

```python
df["category_1"] = df_category["category_1"]
df["category_2"] = df_category["category_2"]
df["category_3"] = df_category["category_3"]
```

```python
df.drop('category', axis=1, inplace=True)
```

```python
#Checking the missing value in columns
df.isna().sum()
```

```python
df['category_1'].value_counts()
```

```python
df['category_1'] = df['category_1'].str.replace("&", " & ")
df['category_1'] = df['category_1'].str.replace("Homelmprovement", "Home lmprovement")
df['category_1'] = df['category_1'].str.replace("Musicallnstruments", "Musical lnstruments")
df['category_1'] = df['category_1'].str.replace("OfficeProducts", "Office Products")
```

```python
df['category_2'].value_counts()
```

```python
df['category_2'] = df['category_2'].str.replace('&', ' & ')
df['category_2'] = df['category_2'].str.replace(',', ', ')
```

```python
df["category_3"].value_counts()
```

```python
#Removing "₹" and "," from 'discounted_price' column and "%" from 'discounted_percentage' column
df["discounted_price"] = df["discounted_price"].str.replace("₹", "")
df["discounted_price"] = df["discounted_price"].str.replace(",", "")
df["discounted_price"] = df["discounted_price"].astype("float32")

df["actual_price"] = df["actual_price"].str.replace("₹", "")
df["actual_price"] = df["actual_price"].str.replace(",", "")
df["actual_price"] = df["actual_price"].astype("float32")

df["discount_percentage"] = df["discount_percentage"].str.replace("%", "")
df["discount_percentage"] = df["discount_percentage"].astype("float32")
```

```python
df["rating"].value_counts()
```

```python
#Replacing "|" in 'rating'  by 4.0
df.query('rating == "|"')
```

```python
df['rating'] = df['rating'].str.replace('|', '4.0').astype('float64')
```

```python
#Removing "," in 'rating_count' column
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
#Splitting the strings in 'user_id' and 'user_name' columns
df["user_id"] = df['user_id'].str.split(',', expand=False)
df['user_name'] = df['user_name'].str.split(',', expand=False)
```

Linear Regression Model

```python
X = df[["rating", "rating_count","actual_price"]]
y = df["discounted_price"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

pred = lr_model.predict(X_test)

print('Linear Regression Intercept: ',lr_model.intercept_)
print('Linear Regression Coefficient: ',lr_model.coef_)
print('R2 Score: ', round(r2_score(y_test, pred),3))
```
